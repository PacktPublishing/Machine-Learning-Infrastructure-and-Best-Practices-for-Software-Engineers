```python
# importing the images library
from datasets import load_dataset, Image

# loading a dataset "food101", or more concretely it's split for training
dataset = load_dataset("food101", split="train")
```

```python
# visualizing the first image
dataset[0]["image"]
```

```python
# now, to check which labels are included, let's just check the structure of the dataset
dataset
```

```python
# and the label for that image, which is the numerical code for the label
# "label" is one of the features in the dataset that we downloaded
dataset[0]['label']
```

```python
# we could very quickly take a look at which labels we have in the dataset
set(dataset['label'])
```

```python
# we can also plot the histogram
# to check the distribution of labels in the dataset
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20,10)

sns.histplot(data=dataset['label'], x=dataset['label'])
```
