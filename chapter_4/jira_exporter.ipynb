# Export software issues /bugs/ from JIRA

This short script demonstrates how to export issues from JIRA. 

The code is based on this article: https://datageeks.medium.com/automate-your-jira-tasks-in-python-bbbcb3145a95

```python
# import the atlassian module to be able to connect to JIRA
from atlassian import Jira

# import pretty print module for nices JSON output
import pprint
```

```python
# create the JIRA instance
# please remember to add the link to the atlassian website and to your credentials
jira_instance = Jira(
    #Url of jira server
    url="https://miroslawstaron.atlassian.net/", 
    #loginuser name
    username='email@domain.com', 
    #password/token
    password='your_token',
    cloud=True
)
```

```python
# get all the tasks/ issues for the project
jql_request = 'project = MLBPB'
issues = jira_instance.jql(jql_request)
```

```python
pp= pprint.PrettyPrinter(indent=1)

# print the issues as a JSON file
pp.pprint(issues)
```
